```python
import glob
import os
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Get the list of files in the directory
direct = glob.glob(r"D:\TAU\Final_Project\final mes\26.6 m_s, 21.9 deg\*")

# Initialize lists to store results
angles2 = []
data_avgs2 = []
data_stds2 = []

# Process each file in the directory
for file_path in direct:
    file_name = os.path.basename(file_path)
    
    with open(file_path, "r") as f:
        df = pd.read_csv(f, sep='\t')
        D = []
        parts = file_name.split(',')
        angle_loc = [i for i, part in enumerate(parts) if 'angle' in part]
        if angle_loc:
            angle = float(parts[angle_loc[0]].split('=')[1])
            angles2.append(angle)

        dt = np.array(df.iloc[:, 0])[1] - np.array(df.iloc[:, 0])[2]
        data_avg = np.mean(np.array(df.iloc[:, 1])[np.array(df.iloc[:, 1])<2])
        data_std = np.std(np.array(df.iloc[:, 1])[np.array(df.iloc[:, 1])<2])

        data_avgs2.append(data_avg)
        data_stds2.append(data_std/np.sqrt(len(np.array(df.iloc[:, 1]))))

```

```python
# Get the list of files in the directory
direct = glob.glob(r"D:\TAU\Final_Project\mes2\*.txt")

# Initialize lists to store results
anglesf = []
data_avgsf = []
data_stdsf = []

# Process each file in the directory
for file_path in direct:
    file_name = os.path.basename(file_path)
    
    with open(file_path, "r") as f:
        df = pd.read_csv(f, sep='\t')
        D = []
        parts = file_name.split(',')
        angle_loc = [i for i, part in enumerate(parts) if 'angle' in part]
        if angle_loc:
            angle = float(parts[angle_loc[0]].split('=')[1])
            anglesf.append(angle)

        dt = np.array(df.iloc[:, 0])[1] - np.array(df.iloc[:, 0])[2]
        data_avg = np.mean(np.array(df.iloc[:, 1])[np.array(df.iloc[:, 1])<2])
        data_std = np.std(np.array(df.iloc[:, 1])[np.array(df.iloc[:, 1])<2])

        data_avgsf.append(data_avg)
        data_stdsf.append(data_std/np.sqrt(len(np.array(df.iloc[:, 1]))))
        
        # Plot the results

plt.plot(anglesf, data_avgsf,".")
plt.xlabel('Angle')
plt.ylabel('Average')
plt.title('Data Averages vs Angle')

plt.show()
```
