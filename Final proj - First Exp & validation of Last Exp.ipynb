```python
"""
Fitting parameters of the classical theory equation: Malus' Law, accoding to estimated values from the visualization section.
"""


from scipy.optimize import curve_fit

def func(x, A, B, C, D):
    y= A*(np.cos((2*np.pi/180)*B*(x-C))**2)+ D
    return y
          
parameters1, covariance1 = curve_fit(func, angles1, data_avgs1
    ,p0=[1.1e-5,1/2,6,0.5e-5]
    ,sigma = data_stds1                            
    )

parameters2, covariance2 = curve_fit(func, angles2, data_avgs2
    ,p0=[1.1e-5,1/2,6,0.5e-5]
    ,sigma = data_stds2
    )
          
parametersf, covariancef = curve_fit(func, anglesf, data_avgsf
#    ,p0=[3.5e-5, 1/2, 0.1 ,0.5e-5]
    ,sigma = data_stdsf
    )
          
          
```
